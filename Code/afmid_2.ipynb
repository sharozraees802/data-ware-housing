```python
DimDt = df[['Order Date']].drop_duplicates().sort_values('Order Date').reset_index(drop = True)
DimDt.shape
```

```python
DimDt['Quarter'] = DimDt['Order Date'].apply(lambda x: x.quarter)
DimDt['Year'] = DimDt['Order Date'].apply(lambda x: x.year)
DimDt['Month'] = DimDt['Order Date'].apply(lambda x: x.month)
DimDt['Week'] = DimDt['Order Date'].apply(lambda x: x.week)
DimDt['Day'] = DimDt['Order Date'].apply(lambda x: x.day)
```

```python
DimDt.head()
```

```python
DtRw = DimDt.shape[0]
DimDt.insert(0, 'OrderDateKey', range(1, 1+DtRw))
DimDt.set_index('OrderDateKey')
DimDt.head()
```

```python
DimDt.head()
```

```python
Salesdf = df[['Order Date', 'City', 'Product ID','Sales', 'Quantity', 'Discount','Profit']]
```

```python
SalesFct = Salesdf.groupby(['Order Date', 'City', 'Product ID']).agg(SalesQtyTotal=('Quantity', 'sum'), SalesAmtTotal = ('Sales', 'sum'), DisAmtTotal=('Discount','sum'), ProfAmtTotal=('Profit','sum')).reset_index()
```

```python
print(SalesFct.shape)
```

```python
SalesFct.head()
```

```python
SalesFct['CostAmt'] = SalesFct['SalesAmtTotal'] - SalesFct['ProfAmtTotal']
```

```python
SalesFct.head()
```

```python
SalesFct.to_csv('G:/DataSets/SalesFct.csv', index=False)
```
